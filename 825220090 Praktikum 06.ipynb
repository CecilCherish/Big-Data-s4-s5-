```python
import matplotlib.pyplot as plt
import networkx as nx
# import library

G = nx.Graph() # membuat graf kosong
G.add_edge("a", "b", weight=0.6)
G.add_edge("a", "c", weight=0.2)
G.add_edge("c", "d", weight=0.1)
G.add_edge("c", "e", weight=0.7)
G.add_edge("c", "f", weight=0.9)
G.add_edge("a", "d", weight=0.3)
elarge = [(u, v) for (u, v, d) in G.edges(data=True) if d["weight"] > 0.5] # Membuat list elarge yang berisi edge-edge dengan weight lebih besar dari 0.5
esmall = [(u, v) for (u, v, d) in G.edges (data=True) if d["weight"] <= 0.5] # dan esmall yang berisi edge-edge dengan weight kurang dari atau sama dengan 0.5
pos = nx.spring_layout(G, seed=7) # Menyusun posisi node-node
# positions for all nodes seed for reproducibility

#nodes
nx.draw_networkx_nodes(G, pos, node_size=700) # menggambarkan garf G, posisi node, ukuran node

# edges
nx.draw_networkx_edges (G, pos, edgelist = elarge, width=6)  # Menggambar node-node graf
nx.draw_networkx_edges(G, pos, edgelist = esmall, width=6, alpha=0.5, edge_color="b", style="dashed")

#labels
nx.draw_networkx_labels(G, pos, font_size=20, font_family="sans-serif") # menggambar graf G, posisi node, ukuran font, jenis font

# edge labels
edge_labels = nx.get_edge_attributes(G, 'weight') # mengambil atribut weight dari setiap edge dan menyimpannya dalam bentuk dictionary
nx.draw_networkx_edge_labels (G, pos, edge_labels = edge_labels) # menggambar label"
ax = plt.gca()   # mendapatkan akses ke objek axis saat ini
ax.margins (0.88)  # mengatur margin dan menyembunyikan axis
plt.axis("off") # menghilangkan sumbu
plt.tight_layout() # mengatur tataletak gambar
plt.show()  # menampilkan
```

```python
import matplotlib.pyplot as plt
import networkx as nx
# import library

# Membuat graf acak geometris dengan 200 node dan peluang koneksi 0.125
G = nx.random_geometric_graph (200, 0.125, seed=896803)

#  Mengambil posisi setiap node dari graf
pos = nx.get_node_attributes (G, "pos")

# Inisialisasi variabel untuk menemukan node sentral
dmin = 1
ncenter = 0

# Mencari node sentral dengan jarak terpendek dari pusat (0.5, 0.5)
for n in pos:
  x, y = pos[n]
  d = (x-0.5) ** 2 + (y - 0.5) ** 2
if d < dmin:
  ncenter = n
  dmin = d

#  Menghitung jalur terpendek dari node sentral ke setiap simpul lainnya
p = dict(nx.single_source_shortest_path_length(G, ncenter))

# Plotting graf dengan warna node berdasarkan jarak dari node sentral
plt.figure(figsize=(8, 8))
nx.draw_networkx_edges (G, pos, alpha=0.4)
nx.draw_networkx_nodes(
    G,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05) # menetapkan sumbu x
plt.ylim(-0.05, 1.05) # menetapkan sumbu y
plt.axis("off") # mematikan axis
plt.show() # menampilkan
```

```python
import numpy as np
from sklearn.metrics.cluster import normalized_mutual_info_score
import pandas as pd
# import library

def DataGraph(dep):
  # Menghitung jumlah baris dan kolom dalam matriks dep
  r1 = dep.shape[0]
  c1 = dep.shape[1]

  # Menetapkan nilai ambang batas (threshold)
  threshold = 0.8

  up = np.triu(dep, k=0) # Membuat matriks segitiga atas dari matriks dep
  var = ["N1", "N2", "N3", "N4", "N5", "N6", "N7", "N8", "N9", "N10", "N11", "N12", "N13", "N14", "N15", "N16", "N17", "N18", "N19", "N20", "N21", "N22", "N23", "N24", "N25"] # Daftar nama variabel
  graph = [] # untuk menyimpan daftar edge
  for i in range(0,r1): # loop untuk mencari edge
    for j in range(0,c1):
      if up[i][j] > threshold:
        x = var[i]
        y = var[j]
        temp = [x,y]
        graph.append(temp)
  return graph #  Mengembalikan daftar edge yang telah dibuat
```

```python
def draw_graph(graph, labels=None, graph_layout = 'shell',
               node_size=1600, node_color='blue', node_alpha=0.2,
               node_text_size=12,
               edge_color='blue', edge_alpha=0.4, edge_thickness=1,
               edge_text_pos=0.3, text_font='sans-serif'):
  # untuk menggambar graf berdasarkan daftar edge

  # membuat graf G baru
  G = nx.Graph()

  # menambahkan edge kedalam loop
  for edge in graph:
    G.add_edge(edge[0], edge[1])

  # layout
  if graph_layout == 'spring':
    graph_pos=nx.spring_layout(G)
  elif graph_layout == 'spectral':
    graph_pos=nx.spextral_layout(G)
  elif graph_layout == 'random':
    graph_pos=nx.random_layout(G)
  else: graph_pos=nx.shell_layout(G)
  # untuk mengatur lay out

  # menggambar graf
  pos = nx.fruchterman_reingold_layout(G) # menentukan posisi node"

  nx.draw_networkx_nodes(G, pos, node_size=node_size,
                         alpha=node_alpha, node_color=node_color) # menggambar graf G, mengatur ukuran, transparansi, dan warna node berdasarkan parameter

  nx.draw_networkx_edges(G, pos, width=edge_thickness,
                         alpha=edge_alpha, edge_color=edge_color) # menggambar graf G, mengatur ketebalan, transparansi, dan warna edge berdasarkan parameter
  nx.draw_networkx_labels(G, pos, font_size=node_text_size,
                          font_family=text_font) # menggambar graf G, mengatur ukuran dan jenis font label

# agar setiap node memiliki label yang berbeda
  if labels is None:
    labels = range(len(graph))

# menampilkan graf
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx
# import library

seed = 13648 # Seed random number generators for reproducibility [ppt]
# Seed digunakan untuk mengatur bilangan acak agar hasilnya dapat direproduksi
G = nx.random_k_out_graph (10, 3, 0.5, seed=seed) # membuat graf acak, 10 jumlah node, 3 jumlah adj dari node, 0.5 probabilitas adj akan terhubung dengan node
pos = nx.spring_layout (G, seed=seed) # menentukan posisi node

node_sizes = [3 + 10 * i for i in range(len(G))] # nentukan ukuran dari setiap node
M = G.number_of_edges() # menyimpan total edge
edge_colors = range(2, M + 2) # menentukan warna
edge_alphas = [(5+1) / (M + 4) for i in range(M)] # menentukan transparansi
cmap = plt.cm.plasma # menentukan color map

nodes = nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color="indigo") # menggambar graf G, ukuran node, warna node
edges = nx.draw_networkx_edges(
    G, # graf G
    pos, # posisi
    node_size=node_sizes, # mengatur ukuran node
    arrowstyle="->",   # mengatur tampilan  panah pada ujung edge
    arrowsize=10,  # mengatur ukuran panah
    edge_color=edge_colors, # warna edge
    edge_cmap=cmap, # colormap yang digunakan
    width=2, # ketebalan
)

# set alpha value for each edge
for i in range(M):
  edges[i].set_alpha(edge_alphas[i])
# ransi dari setiap edge yang digambar sebelumnya disesuaikan berdasarkan
# nilai yang telah ditentukan sebelumnya dalam variabel edge_alphas

pc = mpl.collections.PatchCollection (edges, cmap=cmap) # edge digunakan untuk membuat colormap
pc.set_array(edge_colors) # Menetapkan array warna edge
plt.colorbar(pc) # Menambahkan colorbar ke plot berdasarkan objek PatchCollection

ax = plt.gca() # menyesuaikan atau menambahkan fitur-fitur lainnya ke plot
ax.set_axis_off() # mematikan axis
plt.show() # menampilkan
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import networkx.algorithms.approximation as nx_app
import math
# import library

G = nx.random_geometric_graph (20, radius=0.4, seed=3) # membuat graf acak, dengan 20 node, radius 0.4, seed 3 untuk mengatur bilangan acak
pos = nx.get_node_attributes (G, "pos")


# Depot pada (0,0)
pos[0] = (0.5, 0.5) # mengubah posisi dari simpul dengan indeks 0 menjadi (0.5, 0.5)
H = G.copy() # membuat salinan H

# Menghitung jarak antar node sebagai bobot edge
for i in range(len(pos)): # menyimpan index node menggunakan loop
  for j in range(i + 1, len(pos)):
    dist = math.hypot(pos[i][0] - pos[j] [0], pos[i][1] - pos[j][1])
    dist = dist
    G.add_edge(i, j, weight=dist)
# menambahkan edge antara setiap node dalam graf G dengan weight yang merupakan jarak lurus antara node tersebut berdasarkan posisi node

cycle = nx_app.christofides (G, weight="weight") # untuk menujukan siklus yang melalui semua node
edge_list = list(nx.utils.pairwise (cycle)) #  untuk mengonversi siklus yang telah ditemukan menjadi daftar edge

# Gambarkan edge terdekat pada setiap simpul saja
nx.draw_networkx_edges (H, pos, edge_color="blue", width=0.5)

#menggambarkan rutenya
nx.draw_networkx(
    G, # graf G
    pos, # posisi
    with_labels=True, # menggunakan label
    edgelist=edge_list, # edge list
    edge_color="red", # warna edge
    node_size=200, # ukuran node
    width=3, # ketebalan 3
)

print("The route of the traveller is:", cycle) # mencetak
plt.show() # menampilkan
```
