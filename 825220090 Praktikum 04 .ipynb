```python
import numpy as np # untuk fungsi matematika
import pandas as pd # untuk analisis dan manipulasi data atau menggunakan struktur dataframe
from sklearn import datasets # kumpulan dataset untuk menganalisis data
import matplotlib.pyplot as plt # untuk membuat visualisasi dan plot grafik

#  melakukan import library
```

```python
iris = pd.read_csv('Iris.csv')
# variable iris diisi dengan file Iris.csv
```

```python
import seaborn as sns
ax = sns.boxplot(data=df)
# import library seaborn agar diagram menjadi warna warni/berwarna, sebagai sns
# mengidentifikasi kolom-kolom numerik dalam DataFrame dan membuat boxplot untuk setiap kolomnya
```

```python
tips = sns.load_dataset("tips")
tips.head()
# memuat dataset yang disediakan oleh library seaborn dan menampilkannya dari head/depan/urutan terdepan
```

```python
ax = sns.boxplot(x = "day", y = "total_bill", data = tips)
# mengisi variable ax dengan sumbu x (horizontal) = day dan sumbu y (vertikal) = total_bill untuk diagram boxplot
# data = tips   untuk menentukan dataset yang akan digunakan untuk membuat diagram boxplot

# harusnya beda warna
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
# menentukan sumbu x dan y, memisahkan diagram sesuai dengan data status smoker, dataset menggunakan tips dari library seaborn, dan palette atau warna menggunakan Set3
# dan memasukannnya kedalam variabel ax
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="time", data=tips, linewidth=2.5)
# sumbu x = day, y=total_bill, memisahkan data sesuai dengan data time, data type menggunakan tips dan ketebalan garis = 2.5
```

```python
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))  # fig merupakan variable yang menampung gambar, nrows=2 arti 2 baris,  ncols=3 arti 3 kolom,  figsize=(15, 8) ukuran gambar dalam satuan inch yaitu lebar 15 inci dan tinggi 8 inci
ax[0,0].scatter(df['SepalLengthCm'],df['SepalWidthCm'],color='magenta') # untuk sumbu x = SepalLengthCm dan y = SepalWidthCm, dan warna = magenta
ax[0,0].title.set_text('SepalLength and SepalWidth\n'+'Correlation = ' + "{:.2f}".format(c.iloc[0]['SepalWidthCm'])) #penamaan/penulisan title
ax[0,0].set_xlabel('Sepal Length') # penulisan/penamaan horizontal
ax[0,0].set_ylabel('Sepal Width') # penulisan/penamaan vertikal

# untuk keterangan selanjutnya sama denga yang diatas jadi tidak ditulis ulang

ax[0,1].scatter (df['SepalLengthCm'], df['PetalLengthCm'], color='green')
ax[0,1].title.set_text('SepalLength and PetalLength\n' +'Correlation = ' + "{:.2f}".format(c.iloc[0]['PetalLengthCm']))
ax[0,1].set_xlabel('Sepal Length')
ax[0,1].set_ylabel('Petal Length')

ax[0,2].scatter(df['SepalLengthCm'], df['PetalWidthCm'], color='blue')
ax[0,2].title.set_text('SepalLength and PetalWidth\n' + 'Correlation = ' + "{:.2f}".format(c.iloc[0]['PetalWidthCm']))
ax[0,2].set_xlabel('Sepal Length')
ax[0,2].set_ylabel('Petal Width')

ax[1,0].scatter (df['SepalWidthCm'], df['PetalLengthCm'],color='cyan')
ax[1,0].title.set_text('SepalWidth and PetalLength\n' + 'Correlation = ' + "{:.2f}".format(c.iloc[1] ['PetalLengthCm']))
ax[1,0].set_xlabel('Sepal Width')
ax[1,0].set_ylabel('Petal Length')

ax[1,1].scatter(df['SepalWidthCm'], df['PetalWidthCm'], color='red')
ax[1,1].title.set_text('Sepalwidth and PetalWidth\n' +'Correlation = ' + "{:.2f}".format(c.iloc[1] ['PetalWidthCm']))
ax[1,1].set_xlabel('Sepal Width')
ax[1,1].set_ylabel('Petal Width')

ax[1,2].scatter(df['PetalLengthCm'], df['PetalWidthCm'],color='orange')
ax[1,2].title.set_text('PetalLength and PetalWidth\n'+'Correlation = ' + "{:.2f}".format(c.iloc[2] ['PetalWidthCm']))
ax[1,2].set_xlabel('Petal Length')
ax[1,2].set_ylabel('Petal Width')

fig.subplots_adjust(wspace=0.3, hspace=0.4) # untuk menyesuaikan gambar diagram, wspace(width space/space lebar)=0.3, hspace(height space/tinggi space)=0.4
plt.show()

```
